```python
import cv2
import numpy as np
from google.colab import files
import math

print("Paso 1: Por favor sube tu imagen a continuaci칩n:")
uploaded = files.upload()

nombre_archivo = list(uploaded.keys())[0]

def analizar_imagen_con_entropia(ruta_imagen):
    # Cargamos la imagen
    imagen = cv2.imread(ruta_imagen, cv2.IMREAD_UNCHANGED)

    if imagen is None:
        print("Error al leer la imagen.")
        return

    dimensiones = imagen.shape
    print(f"\n--- REPORTE DE LA IMAGEN: {ruta_imagen} ---")

    # --- 1. Informaci칩n B치sica ---
    if len(dimensiones) == 2:
        print(" TIPO: Escala de Grises (1 canal)")
    elif len(dimensiones) == 3:
        canales = dimensiones[2]
        if canales == 3:
            print(" TIPO: Color RGB/BGR (3 canales)")
        elif canales == 4:
            print(" TIPO: Color con Transparencia (4 canales)")

    print(f"游늺 TAMA칌O: {dimensiones[0]} alto x {dimensiones[1]} ancho")

    # --- 2. C츼LCULO DE ENTROP칈A ---
    # Paso A: "Aplanamos" la imagen. Convertimos la matriz 2D/3D en una fila gigante de n칰meros
    # Esto es necesario para contar la frecuencia de todos los p칤xeles juntos.
    pixels_planos = imagen.flatten()

    # Paso B: Calculamos el histograma (frecuencia de cada valor de p칤xel)
    # values = los valores (0-255), counts = cu치ntas veces aparece cada uno
    values, counts = np.unique(pixels_planos, return_counts=True)

    # Paso C: Normalizamos para obtener probabilidades (p)
    # p = conteo / total_de_pixeles
    probabilidades = counts / len(pixels_planos)

    # Paso D: F칩rmula de Shannon: - sum(p * log2(p))
    # Usamos np.log2 para hacerlo r치pido con vectores
    entropia = -np.sum(probabilidades * np.log2(probabilidades))

    print(f" ENTROP칈A DE SHANNON: {entropia:.4f} bits")
    print("-" * 40)

    # --- 3. Mostrar Matriz ---
    print(">> MATRIZ NUM칄RICA (Resumen):")
    print(imagen)

# Ejecutamos la funci칩n
analizar_imagen_con_entropia(nombre_archivo)
```
