```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
import IPython.display as ipd

print("Sube tu archivo de audio (.wav o .mp3):")
uploaded = files.upload()
nombre_archivo = list(uploaded.keys())[0]

def visualizar_onda(ruta_audio):
    print(f"\nProcesando: {ruta_audio} ...")

    y, sr = librosa.load(ruta_audio)

    plt.figure(figsize=(14, 5))

    librosa.display.waveshow(y, sr=sr, color="blue", alpha=0.6)

    plt.title(f"Forma de onda: {ruta_audio}")
    plt.xlabel("Tiempo (segundos)")
    plt.ylabel("Amplitud")
    plt.grid(True, alpha=0.3)
    plt.show()

    print("Reproducir audio original:")
    ipd.display(ipd.Audio(ruta_audio))

visualizar_onda(nombre_archivo)
```
